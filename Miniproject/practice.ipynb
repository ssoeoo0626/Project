```python
import requests
```

```python
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://finance.yahoo.com'
response = requests.get(url)
```

```python
html_data = response.text
```

```python
soup = bs(html_data, 'html.parser')
```

```python
soup
```

```python
div_data = soup.find('div', attrs={
    'id': 'quote-summary'
})
div_data
```
