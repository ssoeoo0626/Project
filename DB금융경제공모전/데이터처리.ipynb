```python
import pandas as pd
import os

# CSV 파일 읽기
tmp = pd.read_csv('05_02_주요_국제금리.csv')
국가리스트 = tmp.loc[1:, '국가별'].tolist()
tmp.set_index('국가별', inplace=True)

# 국가 리스트 반복
for 국가 in 국가리스트:
    # 해당 국가 데이터 추출
    국가_df = tmp.loc[['국가별', 국가]].transpose().reset_index()
    국가_df['index'] = 국가_df['index'].str.split('.').str[0]

    # '-' 값을 0으로 대체
    국가_df = 국가_df.replace('-', 0)

    # Pivot Table 생성
    국가_df = 국가_df.pivot_table(
        index='index',
        columns='국가별',
        values=국가,
        aggfunc='mean'
    )
    국가_df.columns.name = ''
    국가_df.index.name = ''

    # CSV 파일로 저장
    output_dir = './05_02_output/'
    os.makedirs(output_dir, exist_ok=True)
    국가_df.to_csv(os.path.join(output_dir, f'{국가}_result.csv'))
```

```python
df5_2= pd.read_csv('05_02_주요_국제금리.csv').transpose()
df5_2= df5_2.reset_index()
df5_2
```

```python
df5_2.columns = ['날짜', '금리종류', '대한민국', '중국', '홍콩', '인도', '일본', '싱가포르', '태국', '미국']
df5_2 = df5_2.iloc[1:, :]
```

```python
df5_2['날짜']= df5_2['날짜'].str.split('.').str[0]
df5_2.head(10)
```

```python
countries = ['대한민국', '중국', '홍콩', '인도', '일본', '싱가포르', '태국', '미국']
country_dfs = {} 
output_dir = './05_02_output/'
for country in countries:
    country_df = df5_2[['날짜', '금리종류', country]]
    country_df['날짜'] = country_df['날짜'].str.split('.').str[0]
    country_df['날짜'] = pd.to_datetime(country_df['날짜'])
    
    country_dfs[country] = country_df

    tmp = country_dfs[country].pivot_table(
        index='날짜', 
        columns='금리종류', 
        values=country,
        aggfunc='mean'
    )
    tmp.columns.name = ''

    os.makedirs(output_dir, exist_ok=True)
    tmp.to_csv(os.path.join(output_dir, f'{country}_result.csv'))
```

```python
import pandas as pd

def df7(_df):
    _df = _df[['LOCATION', 'TIME', 'Value']]
    _df.columns = ['국가명', 'Date', 'CLI']
    _df['Date'] = pd.to_datetime(_df['Date'])
    _df.set_index('Date', inplace=True)
    return _df
```

```python
import pandas as pd
import os

# 국가 코드와 파일 경로를 딕셔너리로 정의
countries = {
    'BRA': './07_oecd/BRA_oecd.csv',
    'CAN': './07_oecd/CAN_oecd.csv',
    'CHN': './07_oecd/CHN_oecd.csv',
    'DEU': './07_oecd/DEU_oecd.csv',
    'ESP': './07_oecd/ESP_oecd.csv',
    'FRA': './07_oecd/FRA_oecd.csv',
    'G_7': './07_oecd/G_7_oecd.csv',
    'G_20': './07_oecd/G_20_oecd.csv',
    'GBR': './07_oecd/GBR_oecd.csv',
    'IDN': './07_oecd/IDN_oecd.csv',
    'ITA': './07_oecd/ITA_oecd.csv',
    'JPN': './07_oecd/JPN_oecd.csv',
    'KOR': './07_oecd/KOR_oecd.csv',
    'MEX': './07_oecd/MEX_oecd.csv',
    'TUR': './07_oecd/TUR_oecd.csv',
    'USA': './07_oecd/USA_oecd.csv',
    'ZAF': './07_oecd/ZAF_oecd.csv',
}

# 결과 데이터프레임을 담을 딕셔너리 초기화
dfs = {}

# 반복문을 통해 각 국가의 CSV 파일을 읽어서 데이터프레임 생성 및 저장
for country_code, file_path in countries.items():
    df = pd.read_csv(file_path, index_col=0)
    df = df7(df)  # 위에서 정의한 함수를 사용하여 데이터프레임 가공
    dfs[f'df7_{country_code}'] = df
    output_dir = './07_oecd_processed/'
    os.makedirs(output_dir, exist_ok=True)  # 디렉토리가 없으면 생성
    df.to_csv(os.path.join(output_dir, f'df7_{country_code}.csv'))

# 결과 확인
for key, df in dfs.items():
    print(f"{key}:\n{df}\n")
```
