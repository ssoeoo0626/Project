```python
def get_numbers_from_url(url):
    response = requests.get(url)
    html = response.text
    soup = bs(html, 'html.parser')

    all_numbers = []
    cells_with_numbers = soup.find_all(['td', 'th'])

    for cell in cells_with_numbers:
        try:
            number = float(cell.get_text().replace(',', '').replace('%', ''))  # 숫자 데이터에서 쉼표와 퍼센트 기호 제거 후 숫자로 변환
            all_numbers.append(number)
        except ValueError:
            pass

    return all_numbers
```

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
def practice(ticker):
    url = 'https://navercomp.wisereport.co.kr/v2/company/c1010001.aspx?cmp_cd='
    df = pd.DataFrame([], index = ticker, columns= ['종목명', '업종분류', 'EPS', 'BPS', 'PER', '업종PER', 'PBR', '현금배당수익률'])
    for code in ticker:
        real_url = url + code

        soup = bs(requests.get(real_url).text)

        # kospi
        new_td_list = []
        table_data = soup.find_all('table', attrs={'class' : 'cmp-table'})[0].find_all('td')
        td_list = table_data[0].find_all('dt',{'class':'line-left'})
        for element in td_list:
            text2 = element.get_text().split(':')
            new_td_list.append(text2)
            for i in new_td_list:
                series = pd.Series(i)
                df = df.append(series, ignore_index=True)


        # 밑에꺼
        new_td_list2 = []
        td_list2 = table_data[2].find_all('dt',{'class':'line-left'})
        for element in td_list2:
            text = element.get_text()
            new_td_list2.append(text)

    return df 
```
