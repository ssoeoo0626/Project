```python
from datetime import datetime, timedelta

def print_date_range(start_date, end_date):
    # 시작 날짜와 끝 날짜를 datetime 객체로 변환
    start = datetime.strptime(start_date, "%Y-%m-%d")
    end = datetime.strptime(end_date, "%Y-%m-%d")

    # 시작 날짜부터 끝 날짜까지 날짜 출력
    current_date = start
    while current_date <= end:
        print(current_date.strftime("%Y-%m-%d"))
        current_date += timedelta(days=1)

# 함수 호출하여 날짜 출력
print_date_range('2010-01-01', '2023-09-30')
```

```python
import pandas as pd
from selenium import webdriver
import time
from datetime import datetime, timedelta

def crawling(start_date, end_date):
    driver = webdriver.Chrome()
    datas = {}

    date_range = pd.date_range(start=start_date, end=end_date)
    date_list = date_range.strftime('%Y-%m-%d').tolist()

    for date in date_list:
        url = "https://www.nicerating.com/disclosure/spreadRates.do?strDate=" + date
        driver.get(url)
        time.sleep(0.3)

        print(url)
        print(f'날짜: {date}', end=', ')

        html_data = driver.page_source
        dfs = pd.read_html(html_data)
        df = dfs[-1]

        datas[date] = df

    driver.close()

    result = pd.concat(datas, axis=0)
    return result
```
