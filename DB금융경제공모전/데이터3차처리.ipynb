```python
# 변화율 계산

import pandas as pd

def calculate_percentage_change(df, sell_end_date, redemption_date):
    # 주어진 날짜 범위에 대한 데이터 추출
    subset = df.loc[sell_end_date:redemption_date, 'Adj Close']

    # 변화율(%) 계산
    percentage_change = subset.pct_change() * 100

    return percentage_change

df = pd.DataFrame(df)

result = calculate_percentage_change(df, '', '')
result 

```
